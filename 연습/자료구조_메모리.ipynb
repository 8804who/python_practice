```python
import time

length = 1000

def time_check(func):
    def wrapper():
        temp = [[0 for _ in range(length)]for _ in range(length)]
        start_time = time.perf_counter()
        func(temp)
        end_time = time.perf_counter()

        elapsed_time = end_time - start_time
        return elapsed_time
    return wrapper
```

```python
@time_check
def run(temp):
    for i in range(length):
        for j in range(length):
            temp[i][j] = i+j

total_runtime = 0
for _ in range(10):
    total_runtime += run()
avg_runtime = total_runtime/10
print(f"평균 실행시간: {avg_runtime}")
```
